```python
import numpy as np #Numpy Array
import pandas as pd #load and process data
from sklearn.model_selection import train_test_split #training and testing
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #Accuracy
```

```python
#Separating Data and Labels

x=sonar_data.drop(columns=60, axis=1)
y=sonar_data[60]
```

```python
print('Datas Are:')
print(x)
print('Labels Are:')
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

### Converting Target Labels to Numerical Values

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print('Original labels:', y.unique())
print('Encoded labels:', label_encoder.classes_)
print('First 5 encoded labels:', y_encoded[:5])

y = pd.Series(y_encoded, name='target')
```

Now the `y` variable contains numerical labels (0s and 1s) instead of 'R' and 'M'. This processed `y` will be used for training the model.

### Data Split Visualization

This diagram conceptually illustrates how the original `sonar_data` is first separated into `X` (features) and `y` (labels), and then further split into `x_train`, `y_train`, `x_test`, and `y_test` for machine learning.

-   **Original Data:** The entire dataset.
-   **Features (X):** The input measurements (columns 0-59).
-   **Labels (y):** The target variable (column 60, 'R'/'M', converted to 0/1).
-   **Training Set:** Data used to teach the model (`x_train`, `y_train`).
-   **Testing Set:** Data used to evaluate the model on unseen examples (`x_test`, `y_test`).

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 8))
ax.set_aspect('equal', adjustable='box')

# Original Data
original_data_rect = patches.Rectangle((0, 0), 10, 8, linewidth=1, edgecolor='black', facecolor='lightgray', label='Original Data')
ax.add_patch(original_data_rect)
ax.text(5, 4, 'Original sonar_data (208 rows, 61 columns)', ha='center', va='center', fontsize=12, weight='bold')

# Split into X and y
ax.text(11, 4, '--> Separate features (X) and labels (y)', ha='left', va='center', fontsize=10)

x_rect = patches.Rectangle((15, 5), 6, 2, linewidth=1, edgecolor='blue', facecolor='skyblue', label='Features (X)')
ax.add_patch(x_rect)
ax.text(18, 6, 'X (208 rows, 60 columns)', ha='center', va='center', fontsize=10)

y_rect = patches.Rectangle((15, 1), 6, 2, linewidth=1, edgecolor='red', facecolor='lightcoral', label='Labels (y)')
ax.add_patch(y_rect)
ax.text(18, 2, 'y (208 rows, 1 column)', ha='center', va='center', fontsize=10)

# Train-Test Split
ax.text(22, 4, '--> Train-Test Split (e.g., 90% train, 10% test)', ha='left', va='center', fontsize=10)

x_train_rect = patches.Rectangle((27, 6), 4, 1.5, linewidth=1, edgecolor='darkblue', facecolor='lightblue', label='x_train')
ax.add_patch(x_train_rect)
ax.text(29, 6.75, 'x_train (187 rows, 60 cols)', ha='center', va='center', fontsize=8)

y_train_rect = patches.Rectangle((27, 4), 4, 1.5, linewidth=1, edgecolor='darkred', facecolor='salmon', label='y_train')
ax.add_patch(y_train_rect)
ax.text(29, 4.75, 'y_train (187 rows, 1 col)', ha='center', va='center', fontsize=8)

x_test_rect = patches.Rectangle((27, 2), 4, 1.5, linewidth=1, edgecolor='darkgreen', facecolor='lightgreen', label='x_test')
ax.add_patch(x_test_rect)
ax.text(29, 2.75, 'x_test (21 rows, 60 cols)', ha='center', va='center', fontsize=8)

y_test_rect = patches.Rectangle((27, 0), 4, 1.5, linewidth=1, edgecolor='darkgoldenrod', facecolor='gold', label='y_test')
ax.add_patch(y_test_rect)
ax.text(29, 0.75, 'y_test (21 rows, 1 col)', ha='center', va='center', fontsize=8)

ax.set_xlim(-2, 35)
ax.set_ylim(-2, 10)
ax.axis('off')

plt.title('Conceptual Diagram of Data Splitting', fontsize=14, pad=20)
plt.show()
```

Training Model -> Logistic Regression

```python
# Training the Logistic Regression Model with training data
model = LogisticRegression()
```

```python
# Training the model

model.fit(x_train,y_train)
```

```python
predict_x_train = model.predict(x_train)
training_data_accuracy = accuracy_score(predict_x_train,y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
predict_x_test = model.predict(x_test)
test_data_accuracy = accuracy_score(predict_x_test,y_test)
```

```python
print('Accuracy on testing data: ',test_data_accuracy)
```
